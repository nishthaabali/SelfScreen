# Spliting Features and Target

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

# Splitting data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=5)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model= KNeighborsClassifier(n_neighbors=3)
```

```python
# training the Logistic Regression model using Training data

model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data using SVM= {:.2f}%'.format(training_data_accuracy * 100))
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data using SVM= {:.2f}%'.format(test_data_accuracy * 100))
```

```python
from sklearn.metrics import precision_score

# Assuming y_train contains the true labels and y_pred_train contains the predicted labels on the training data

# Calculate precision on the training data
X_train_prediction = model.predict(X_train)
test_data_accuracy = precision_score(Y_train, X_train_prediction)
print('Precision on training data using SVM= {:.2f}%'.format(test_data_accuracy * 100))
```
